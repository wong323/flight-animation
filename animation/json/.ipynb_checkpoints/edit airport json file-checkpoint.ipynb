# Filter airport json file
This script is to read in json file and only keep airports we want to be displayed

```python
import json
import pandas as pd
```

```python
ups_file_path =r'S:\Analytics\4 - Analytics Consulting & Visualization\Professional Services Consulting Support\UPS\7478F Inventory\Analysis\Code Repo\data\interim\csvs\planned_flights.csv'
df = pd.read_csv(ups_file_path)
```

```python
list_airports = list(set(df['ORIG'])) + list(set(df['DEST']))
list_airports = list(set(list_airports))
```

```python
file = 'airports.topo.json'
with open(file) as train_file:
    dict_train = json.load(train_file)
```

```python
len(dict_train['objects']['airports']['geometries'])
```

```python
count = 0
matched_list = []
res = []
for i in range(0, len(dict_train['objects']['airports']['geometries'])):
    x = dict_train['objects']['airports']['geometries'][i]['id']
    
    if x in list_airports:
        count+=1
        matched_list.append(x)
        res.append(dict_train['objects']['airports']['geometries'][i])
    else:
        pass
    
print(len(res))
#print(len(dict_train['objects']['airports']['geometries']))
```

```python
dict_train['objects']['airports']['geometries'] = res
```

# below takes flight schedule and make each od pair into a list of list

```python
df.head()
```

```python
list_of_OD = []
for index, row in df.iterrows():
    orig = row['ORIG']
    dest = row['DEST']
    
    if orig in matched_list:
        if dest in matched_list:
            list_of_OD.append([orig, dest])
```

```python
with open('list_flight_OD.json', 'w') as fp:
    json.dump(list_of_OD, fp)
```
